```python
# Question 4
# We need to convert data to the useful type of data for future researches
```

```python
# First, changing data type

data_train.loc[:, 'Age'] = data_train.loc[:, 'Age'].replace('[\_,]', '', regex = True).astype(int)

data_train.loc[:, 'Annual_Income'] = data_train.loc[:, 'Annual_Income'].replace('[\_,]', '', regex = True).astype(float)

data_train.loc[:, 'Num_of_Loan'] = data_train.loc[:, 'Num_of_Loan'].replace('[\_,]', '', regex = True).astype(int)

imputer_1 = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
data_train['Num_of_Delayed_Payment'] = imputer_1.fit_transform(data_train['Num_of_Delayed_Payment'].array.reshape(-1, 1))
data_train.loc[:, 'Num_of_Delayed_Payment'] = data_train.loc[:, 'Num_of_Delayed_Payment'].replace('[\_,]', '', regex = True).astype(int)
```

```python
imputer_2 = SimpleImputer(missing_values = '', strategy = 'most_frequent')
data_train.loc[:, 'Changed_Credit_Limit'] = data_train.loc[:, 'Changed_Credit_Limit'].replace('[\_,]', '', regex = True)
data_train['Changed_Credit_Limit'] = imputer_2.fit_transform(data_train['Changed_Credit_Limit'].array.reshape(-1, 1))
data_train.loc[:, 'Changed_Credit_Limit'] = data_train.loc[:, 'Changed_Credit_Limit'].astype(float)

data_train.loc[:, 'Outstanding_Debt'] = data_train.loc[:, 'Outstanding_Debt'].replace('[\_,]', '', regex = True).astype(float)

data_train.loc[:, 'Amount_invested_monthly'] = data_train.loc[:, 'Amount_invested_monthly'].replace('[\_,]', '', regex = True).astype(float)

data_train.loc[:, 'Monthly_Balance'] = data_train.loc[:, 'Monthly_Balance'].replace('[\_,]', '', regex = True).astype(float)
```

```python
imputer_3 = SimpleImputer(strategy = 'most_frequent')
data_train['Credit_History_Age'] = imputer_3.fit_transform(data_train['Credit_History_Age'].array.reshape(-1, 1))

for i in range(len(data_train['Credit_History_Age'])):
    # find numbers
    s = [int(s) for s in re.findall(r'-?\d+\.?\d*', data_train.iloc[i]['Credit_History_Age'])]
    #print(s[0])
    # replacement
    data_train.loc[i, 'Credit_History_Age'] = s[0]*12 + s[1]
    
data_train.loc[:, 'Credit_History_Age'] = data_train.loc[:, 'Credit_History_Age'].astype(int)
```

```python
# Age Set between 0 and 100
for i in range(len(data_train['Age'])):
    # find numbers
    if (data_train.iloc[i]['Age'] < 0) or (data_train.iloc[i]['Age'] > 100):
        #print(data_train.iloc[i]['Age'])
        data_train.loc[i, 'Age'] = np.nan
    
imputer_10 = SimpleImputer(strategy = 'median')
data_train['Age'] = imputer_10.fit_transform(data_train['Age'].array.reshape(-1, 1))
```

```python
# SSN Clear '#F%$D@*&8'
for i in range(len(data_train['SSN'])):
    # find #F%$D@*&8
    if data_train.iloc[i]['SSN'] == '#F%$D@*&8':
        #print(data_train.iloc[i]['SSN'])
        data_train.loc[i, 'SSN'] = np.nan
    
imputer_11 = SimpleImputer(strategy = 'most_frequent')
data_train['SSN'] = imputer_11.fit_transform(data_train['SSN'].array.reshape(-1, 1))
```

```python
# Occupation Clear '_______'

for i in range(len(data_train['Occupation'])):
    # find #F%$D@*&8
    if data_train.iloc[i]['Occupation'] == '_______':
        #print(data_train.iloc[i]['Occupation'])
        data_train.loc[i, 'Occupation'] = np.nan
    
imputer_12 = SimpleImputer(strategy = 'most_frequent')
data_train['Occupation'] = imputer_12.fit_transform(data_train['Occupation'].array.reshape(-1, 1))
```

```python
# Num_Bank_Accounts Lets set num bank accounts between 1 and 20.

for i in range(len(data_train['Num_Bank_Accounts'])):
    # find numbers
    if (data_train.iloc[i]['Num_Bank_Accounts'] < 1) or (data_train.iloc[i]['Num_Bank_Accounts'] > 20):
        #print(data_train.iloc[i]['Num_Bank_Accounts'])
        data_train.loc[i, 'Num_Bank_Accounts'] = np.nan
    
imputer_13 = SimpleImputer(strategy = 'median')
data_train['Num_Bank_Accounts'] = imputer_13.fit_transform(data_train['Num_Bank_Accounts'].array.reshape(-1, 1)).astype(int)


```

```python
# Num_Credit_Card Set between 1 and 100.

for i in range(len(data_train['Num_Credit_Card'])):
    # find numbers
    if (data_train.iloc[i]['Num_Credit_Card'] < 1) or (data_train.iloc[i]['Num_Credit_Card'] > 100):
        #print(data_train.iloc[i]['Num_Credit_Card'])
        data_train.loc[i, 'Num_Credit_Card'] = np.nan
    
imputer_14 = SimpleImputer(strategy = 'median')
data_train['Num_Credit_Card'] = imputer_14.fit_transform(data_train['Num_Credit_Card'].array.reshape(-1, 1)).astype(int)
```

```python
# Delay_from_due_date Set more than 0

for i in range(len(data_train['Delay_from_due_date'])):
    # find numbers
    if (data_train.iloc[i]['Delay_from_due_date'] < 0):
        #print(data_train.iloc[i]['Delay_from_due_date'])
        data_train.loc[i, 'Delay_from_due_date'] = np.nan
    
imputer_17 = SimpleImputer(strategy = 'median')
data_train['Delay_from_due_date'] = imputer_17.fit_transform(data_train['Delay_from_due_date'].array.reshape(-1, 1)).astype(int)
```

```python
# Num_Credit_Inquiries Set between 0 and 100

for i in range(len(data_train['Num_Credit_Inquiries'])):
    # find numbers
    if (data_train.iloc[i]['Num_Credit_Inquiries'] < 0) or (data_train.iloc[i]['Num_Credit_Inquiries'] > 100):
        #print(data_train.iloc[i]['Num_Credit_Inquiries'])
        data_train.loc[i, 'Num_Credit_Inquiries'] = np.nan
    
imputer_19 = SimpleImputer(strategy = 'median')
data_train['Num_Credit_Inquiries'] = imputer_19.fit_transform(data_train['Num_Credit_Inquiries'].array.reshape(-1, 1)).astype(int)
```
