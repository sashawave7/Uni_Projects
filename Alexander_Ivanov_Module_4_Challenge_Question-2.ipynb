# Module 4 Challenge Question: Olympic Sports and Medals, 1896-2014

data source: https://www.kaggle.com/the-guardian/olympic-games

More than 35,000 medals have been awarded at the Olympics since 1896. This dataset includes a row for every Olympic athlete that has won a medal since the first games.

Using Pandas library,

- Find out the (male and female) athlete who won most medals in all the Summer Olympic Games (1896-2014). 

- Create a barplot which shows the total medals won for each sport during the Summer Olympic Games (1896-2014). 



```
import pandas as pd
import matplotlib.pyplot as plt

summer = pd.read_csv("/content/sample_data/summer.csv")

# First Part
def most_medals_men():
  medals = summer[summer['Gender'] == 'Men'].loc[:, 'Athlete']
  D = dict()
  num_medals = 0
  athlet = None
  for i in medals:
    D[i] = D.get(i, 0) + 1
    if D[i] > num_medals:
      num_medals = D[i]
      athlet = i
    
  return [athlet, num_medals]

def most_medals_women():
  medals = summer[summer['Gender'] == 'Women'].loc[:, 'Athlete']
  D = dict()
  num_medals = 0
  athlet = None
  for i in medals:
    D[i] = D.get(i, 0) + 1
    if D[i] > num_medals:
      num_medals = D[i]
      athlet = i
    
  return [athlet, num_medals]

WinMan = most_medals_men()
WinWoman = most_medals_women()

print('Part 1')
print('Athlete man with most medals (1896-2014):', WinMan[0], 'with', WinMan[1], 'medals')
print('Athlete woman with most medals (1896-2014):', WinWoman[0], 'with', WinWoman[1], 'medals')
print('\n')

#Second Part

# Y - Axis
medals_per_sport = list()
for i in summer['Sport'].unique():
  sporttable = summer.loc[summer['Sport'] == i]
  medals_per_sport.append(len(sporttable))

# X - Axis 
sportlst = list() 
for i in summer['Sport'].unique():
  sportlst.append(i)

#Drawing
print('Part 2')
plt.figure(figsize=(17,10))
plt.bar(sportlst, medals_per_sport)
plt.xticks(rotation= 'vertical', fontsize = 13)
plt.xlabel("Sports")
plt.ylabel("Number of Medals")
plt.title("Medals per sport (1896-2014)") 
plt.show()
```
